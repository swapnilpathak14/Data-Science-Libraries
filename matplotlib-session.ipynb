# Matplotlib
Matplotlib is a desktop plotting package designed for creating (mostly twodimensional) publication-quality plots
1. Matplotlib is a Python package for 2D plotting that generates production-quality graphs.
2. It supports interactive and non-interactive plotting, and can save images in several output formats (PNG, PS, and others)
3. It provides a wide variety of plot types (lines, bars, pie charts, histograms, and many more).
4. The dual nature of Matplotlib allows it to be used in both interactive and non-interactive scripts.

## Why Matplotlib is so successful
1. It uses Python
2. It's open source, so no license to pay
3. It's a real programming language
4. Python has a lot of external modules that will help us perform all the functions we need to. So it's the perfect tool to acquire data, elaborate the data, and then plot the data
5. It's very customizable and extensible
6. It's integrated with LaTeX markup used for scientific papers
7. It's cross-platform and portable

### Multiline Plot
plots are reset after each cell is evaluated, so for more complex plots you must put all of the plotting commands in a single notebook cell<br>

```python
x = np.arange(0.0, 6.0, 0.01)
plt.plot(x, [x**2 for x in x])
plt.plot(x, [x**2.35 for x in x])
plt.plot(x, [x**2.2 for x in x])
```

### Add Legend
1. Add extra keyword argument, `label`, to the `plot()` call
2. call the `legend()` function with no argument
<br>**Important** <br>
Each plot function for each legend should be called seperately
1. `plot()` command keeps the track of the line represents, so that information should be there.
2. The order of lines plotted is important because the legend() parameters have to be matched with the lines

```python
x = np.arange(0.0, 6.0, 0.01)
plt.plot(x, [x**2 for x in x], label='Slow')
plt.plot(x, [x**2.25 for x in x], label='Normal')
plt.plot(x, [x**2.5 for x in x], label='Fast')
plt.grid(True)
plt.legend()
```

Options for Legend Position to be used with `plt.legend(loc='XXX')`

### Markers and Line Styles
`plot()` supports an optional third argument for each pair of X, Y arguments:
`plt.plot(X, Y, '<format>', ...)`

Three levels of customization:
1. Colors
2. Line styles
3. Marker styles

### Color
List of Color abbrevation and color names<br>
`b` blue<br>
`c` cyan<br>
`g` green<br>
`k` black<br>
`m` magenta<br>
`r` red<br>
`w` white<br>
`y` yellow<br>

Ways to specify color<br>
1. The full color name, such as yellow
2. Hexadecimal string (the same format as in HTML code) such as #FF00FF
3. RGB or RGBA tuples, for example, (1, 0, 1, 1)
4. Grayscale intensity, in string format such as '0.7'

```python
x = np.arange(0.0, 6.0, 0.01)
plt.plot(x, [x**2 for x in x], 'y', label='Slow')
plt.plot(x, [x**2.25 for x in x],'m', label='Normal')
plt.plot(x, [x**2.5 for x in x], 'r', label='Fast')
plt.grid(True)
plt.legend(loc='center left')
```

### Pie Chart

```python
plt.figure(figsize=(6,6))
x = [45, 35, 20]
labels = ['Cats', 'Dogs', 'Fishes']
plt.pie(x, labels=labels);
```

Other important parameters

`explode`: Specify the offset radius fraction the wedge has from the center of the pie<br>
`colors`: This is a list of Matplotlib colors, cyclically used to color the wedges.<br>
`labels`, `labeldistance`: This is a list of labels<br>
`autopct` : Label wedges with their numeric values.<br>
`pctdistance`: pctdstance is the ratio between the center of the pie and the start of the text.<br>
`shadow`: This draws a shadow for wedges or pie.

```python
plt.figure(figsize=(3,3))
x = [4, 9, 21, 55, 30, 18]
labels = ['Swiss', 'Austria', 'Spain', 'Italy', 'France', 'Benelux']
explode = [0, 0, 0, 0, 0.1, 0]
plt.pie(x, labels=labels, explode=explode, autopct='%1.1f%%')
```

### Text Inside Figure

The `text()` function writes a string (text) at an arbitrary position (specified by (x,y)):

```python
x = np.arange(0, 2*np.pi, .01)
y = np.sin(x)
plt.plot(x, y)
plt.text(0.1, -0.04, 'sin(0)=0')
```

## Annotations

In annotation, there are two points
1. the graph point we want to annotate (represented by an xy keyword argument) 
2. the plot position where we want to place the annotation text (represented by xytext)
<br>
Both are expressed in an (x,y) format in data coordinate positions

```python

import numpy as np
import matplotlib.pyplot as plt
y = [13, 11, 13, 12, 13, 10, 30, 12, 11, 13, 12, 12, 12, 11, 12]
plt.plot(y)
plt.axis(ymax=35)
plt.annotate('this spot must really\nmean something', xy=(6, 30), xytext=(8, 31.5),
             arrowprops=dict(facecolor='black',shrink=0.05))
```

## Figure and Subplots

Plots in matplotlib reside within a `Figure` object. You can create a new figure with `plt.figure`

`plt.figure` has a number of options<br>
`figsize` will specify the size and aspect ratio if saved to disk<br>
You canâ€™t make a plot with a blank figure<br>
You have to create one or more subplots using `add_subplot`

figure should be 2 Ã— 2 (two rows and two column - total four)<br>
Third parameter 1 means that we are selecting the first of four subplots<br>

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
plt.plot(np.random.randn(50).cumsum(), 'k--') #Matplotlib draws on last subplot
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

### Plotting with Pandas

In pandas multiple columns of data is availabe, along with row and column labels.<br>
Pandas has built-in methods to simplify visualizations from Dataâ€Frame and Series objects

```python
import pandas as pd
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10))
df.plot()
```

Series plot method arguement

`label` - Label for plot legend<br>
`ax` - matplotlib subplot object to plot on; if nothing passed, uses active matplotlib subplot<br>
`style` - Style string, like 'ko--', to be passed to matplotlib<br>
`alpha` - The plot fill opacity (from 0 to 1)<br>
`kind` - Can be 'area', 'bar', 'barh', 'density', 'hist', 'kde', 'line', 'pie'<br>
`logy` - Use logarithmic scaling on the y-axis<br>
`use_index` - Use the object index for tick labels<br>
`rot` - Rotation of tick labels (0 through 360)<br>
`xticks` - Values to use for x-axis ticks<br>
`yticks` - Values to use for y-axis ticks<br>
`xlim` - x-axis limits (e.g., [0, 10])<br>
`ylim` - y-axis limits<br>
`grid` - Display axis grid (on by default)<br>

DataFrame Plot Arguement<br><br>
`subplots` - Plot each DataFrame column in a separate subplot<br>
`sharex` - If subplots=True, share the same x-axis, linking ticks and limits<br>
`sharey` - If subplots=True, share the same y-axis<br>
`figsize` - Size of figure to create as tuple<br>
`title` - Plot title as string<br>
`legend` - Add a subplot legend (True by default)<br>
`sort_columns` - Plot columns in alphabetical order; by default uses existing column order<br>

```python
fig, axes = plt.subplots(2, 1)
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0], color='k', alpha=0.7)
data.plot.barh(ax=axes[1], color='k', alpha=0.7)
```
